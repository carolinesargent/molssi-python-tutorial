## Processing Multiple Files and Writing Files
This lesson will explain importing a python library, using python library functions, processing multiple files using a for loop, and printing output to a new text file.

```python
# Importing library
# import library_name
# output = library_name.function_name(input)
```

```python
# Specify where files are located and get all output files (with .out extension)
file_location = os.path.join('data','outfiles', '*.out')
print(file_location)
```

```python
# Use function glob in the library glob to create list of all filenames that fit the pattern specificed in the input. 
# Here, the input is the file location.
import glob 
filenames = glob.glob(file_location)
print(filenames)
```

```python
for f in filenames:
    outfile = open(f, 'r')
    data = outfile.readlines()
    outfile.close()
    for line in data:
        if 'Final Energy' in line:
            energy_line = line
            words = energy_line.split()
            energy = float(words[3])
            print(energy)
```

```python
# Use os.path.basename to get the name of the file where information came from.
first_file = filenames[0]
print(first_file)

file_name = os.path.basename(first_file)
print(file_name)
```

```python
# Extract molecule name from example above.
name_line = file_name.split('.')
print(name_line[0])
```

```python
# Modifying loop so we can get file names along with each respective energy value
for f in filenames:
    # Get the molecule name
    file_name = os.path.basename(f)
    split_filename = file_name.split('.')
    molecule_name = split_filename[0]
    
    # Read the data
    outfile = open(f,'r')
    data = outfile.readlines()
    outfile.close()
    
    # Loop through the data
    for line in data:
        if 'Final Energy' in line:
            energy_line = line
            words = energy_line.split()
            energy = float(words[3])
            print(molecule_name, energy)
                        
```

```python
# Printing to a file
# filehandle = open('file_name.txt', 'w+')
# w: open a file for writing
# w+: open the file for writing and if the file does not exist, create it
# a: append to existing file
# w+ will overwrite a file if it already exists, a+ will keep what is already there and add additional text to the file
```

```python
datafile = open('energies.txt', 'w+') # This opens the file for writing
for f in filenames:
    # Get the molecule name
    file_name = os.path.basename(f)
    split_filename = file_name.split('.')
    molecule_name = split_filename[0]
    
    # Read the data
    outfile = open(f, 'r')
    data = outfile.readlines()
    outfile.close()
    
    # Loop through the data
    for line in data:
        if 'Final Energy' in line:
            energy_line = line
            words = energy_line.split()
            energy = float(words[3])
            datafile.write(F'{molecule_name} \t {energy} \n') #to convert to string, place a capital F in front 
            # of the line you want to print and enclose it in single quotes. Each python variable is placed in 
            # braces. \t inserts a tab
datafile.close()
```

```python
# Geometry Analysis Project with Extensions
import os
import numpy

file_location = os.path.join('data','water.xyz')

xyz_file = numpy.genfromtxt(fname = file_location, skip_header=2, dtype='unicode')
output_file = open('bond_lengths.txt', 'w+')
# Must use skip header so that numbers of columns are the same.
atom_labels = xyz_file[:,0]
coords = xyz_file[:,1:]
coords = coords.astype(numpy.float64)
num_atoms = len(atom_labels)


for num1 in range(0,num_atoms): #starts with oxygen as num1
    for num2 in range(0, num_atoms): #starts with h1 as num2
        if num1 < num2: #to remove duplicate coordinates
            x_dist = coords[num1,0] - coords[num2,0]
            y_dist = coords[num1,1] - coords[num2,1]
            z_dist = coords[num1,2] - coords[num2,2]
            bond_length = numpy.sqrt(x_dist**2 + y_dist**2 + z_dist**2)
            if bond_length <= 1.5 and bond_length != 0:
                output_file.write(F'{atom_labels[num1]} to {atom_labels[num2]} : {bond_length:.3f}\n')
output_file.close()
```

```python
# Key Points
# If you are reading in a file that is mostly numerical data, there are better ways to read in the data than using
# the readlines()
# The notation to refer to a particular element of an array is array_name[row,column]
# Typically, you should not write a function to perform a standard math operation. The function probably
# already exists in numpy. 
```

## Plotting and Data Visualization
Here, we will plot data to visualize results, label plot axes, create a figure legent, plot multiple graphs on one figure, and save figures to files.

```python
# Getting data 
import os
import numpy

distance_file = os.path.join('data', 'distance_data_headers.csv')
distances = numpy.genfromtxt(fname=distance_file, delimiter=',', dtype='unicode')
headers = distances[0]
data = distances[1:]
data = data.astype(numpy.float64)

# Plotting Data
import matplotlib.pyplot as plt # import library to plot data
plt.figure() # this initializes a new figure
plt.plot(data[:,1])
```

```python
# Labeling plots and saving figures
sample = headers[1]

plt.figure()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstrom)')
fig_1 = plt.plot(data[:,1], label=sample)
plt.legend()

plt.savefig(F'{sample}.png', dpi=300)
```

```python
# Plotting more than one set of data
plt.figure()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstrom)')
plt.plot(data[:,1], label=headers[1])
plt.plot(data[:,2], label=headers[2])
plt.legend()
plt.savefig('two_samples.png')
```

```python
# To plot all samples on the same plot, we can use a for loop to loop through all the columns.

for col in range(1, len(headers)):
    fig = plt.plot(data[:,col], label=headers[col])
    plt.legend()
    
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstrom)')
```

```python
# Writing help messages for our functions
def calculate_distance(atom1_coord, atom2_coord):
    """Calculate the distance between two three-dimensional points."""
    
    x_distance = atom1_coord[0] - atom2_coord[0]
    y_distance = atom1_coord[1] - atom2_coord[1]
    z_distance = atom1_coord[2] - atom2_coord[2]
    bond_length_12 = numpy.sqrt(x_distance ** 2 + y_distance ** 2 + z_distance ** 2)
    return bond_length_12
```

```python
# Function Default Arguments
def bond_check(atom_distance, minimum_length=0, maximum_length=1.5):
    """
    Check if a distance is a bond based on a minimum and maximum bond length.
    """
    
    if atom_distance > minimum_length and atom_distance <= maximum_length:
        return True
    else:
        return False
```
